```python
df_new = df[['ctfp','cwtfp','rwtfpna','csh_g','csh_r','pl_i', 'pl_g', 'pl_x', 'total','excl_energy', 'metals_minerals', 'agriculture', 'total_change', 'fish_change','growthbucket']]
df_new.head()
```

```python
X = df_new.drop('growthbucket', axis=1)
y = df_new[['growthbucket']]
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler instance
scaler = StandardScaler()

# Assuming 'X' is your pandas DataFrame, you can standardize only columns 8 to 11
X.iloc[:, 8:12] = scaler.fit_transform(X.iloc[:, 8:12])
```

```python
y
```

```python
zeroCount = 0
oneCount = 0
for i in y.values:
    if(i==1):
        oneCount += 1
    else:
        zeroCount += 1
```

```python
print(zeroCount)
print(oneCount)
```

```python
import pandas as pd
from imblearn.over_sampling import SMOTE

# Assuming you have your dataset with features 'X' and target variable 'y'

# Initialize the SMOTE object
smote = SMOTE(sampling_strategy='auto', random_state=42)  # You can adjust 'sampling_strategy' as needed

# Apply SMOTE to your dataset
X_resampled, y_resampled = smote.fit_resample(X, y)

# X_resampled and y_resampled now contain the balanced dataset
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
# Initializing the Random Forest classifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
accuracy = rf.score(X_test, y_test)
```
