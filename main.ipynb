```python
df_new = df[['ctfp','cwtfp','rwtfpna','csh_g','csh_r','pl_i', 'pl_g', 'pl_x', 'total','excl_energy', 'metals_minerals', 'agriculture', 'total_change', 'fish_change','growthbucket']]
df_new.head()
```

```python
X = df_new.drop('growthbucket', axis=1)
y = df_new[['growthbucket']]
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler instance
scaler = StandardScaler()

# Assuming 'X' is your pandas DataFrame, you can standardize only columns 8 to 11
X.iloc[:, 8:12] = scaler.fit_transform(X.iloc[:, 8:12])
```

```python
X
```

```python
y
```

```python
zeroCount = 0
oneCount = 0
for i in y.values:
    if(i==1):
        oneCount += 1
    else:
        zeroCount += 1
```

```python
print(zeroCount)
print(oneCount)
```

```python
import pandas as pd
from imblearn import over_sampling
from imblearn.over_sampling import SMOTE

# Assuming you have your dataset with features 'X' and target variable 'y'

# Initialize the SMOTE object
smote = SMOTE(sampling_strategy='auto', random_state=42)  # You can adjust 'sampling_strategy' as needed

# Apply SMOTE to your dataset
X_resampled, y_resampled = smote.fit_resample(X, y)

# X_resampled and y_resampled now contain the balanced dataset
```

```python
#Exloratory Data Analysis
import seaborn as sns
import matplotlib.pyplot as plt

# Create a synthetic dataset


# Calculate the correlation matrix
correlation_matrix = df_new.corr()

# Display the correlation matrix
print(correlation_matrix)

# Visualize the correlation matrix as a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
#Feature Importance Graph
df1 = df_new.copy()
df1.head()

X = df1.drop('growthbucket', axis=1)
y = df1['growthbucket']

from sklearn.ensemble import RandomForestClassifier
# Initializing the Random Forest classifier
rf = RandomForestClassifier()
rf.fit(X, y)
importances = rf.feature_importances_

feature_importances = pd.DataFrame({'Feature': X.columns, 'Importance': importances})
feature_importances = feature_importances.sort_values('Importance', ascending=False)

# Plotting the feature importances
plt.figure(figsize=(10, 6))
plt.barh(feature_importances['Feature'], feature_importances['Importance'])
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importance')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
# Initializing the Random Forest classifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
accuracy = rf.score(X_test, y_test)
```

```python
accuracy
```

```python
from sklearn.metrics import classification_report, confusion_matrix
classification_report(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
#More graphs to show the accuracy of the model
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import random

# Compute ROC curve and ROC area for each class
fpr, tpr, _ = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')  # random predictions curve
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate or (1 - Specifity)')
plt.ylabel('True Positive Rate or (Sensitivity)')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
# Graphing the confusion matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting the confusion matrix
plt.figure(figsize=(10, 6))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, cmap='Blues', fmt='g')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Making more graphs to prove the accuracy of the model
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import average_precision_score

# Compute Precision-Recall and plot curve
precision, recall, _ = precision_recall_curve(y_test, y_pred)
average_precision = average_precision_score(y_test, y_pred)

# Plot Precision-Recall curve
plt.figure(figsize=(10, 6))
plt.plot(recall, precision, label='Precision-Recall curve (area = %0.2f)' % average_precision)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
#More graphs to visualize the accuracy of the model
#Pi chart
import matplotlib.pyplot as plt

# Data to plot
labels = '0', '1'
sizes = [zeroCount, oneCount]
colors = ['gold', 'yellowgreen']
explode = (0.1, 0)  # explode 1st slice

# Plot
plt.figure(figsize=(10, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Pi Chart')
plt.axis('equal')
plt.show()
```
